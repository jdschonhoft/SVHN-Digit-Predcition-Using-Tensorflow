**Answer:**

I did this to limit the number of labels.  Therefore the bounding boxes represent coordinates on a ~6 x 6 grid in the resized images that are 64 x 64.  This model actually performs pretty well.  I also limited the number of convulutiional layers to 4 so that I could train it faster.  The rest was the same.
